```python
import re
```

```python
fr = open("./datasets/out-domain.txt", "r")
text = fr.readlines()
fr.close()
```

```python
def clean(line):
    line = line.lower()
    line = re.sub(r"^(?:what|which|where|who|when|whose|why|how)\s+\w+\s+", "", line)
    line = line.strip()
    
    return line
```

```python
# remove 
cleaned = [clean(line) for line in text]

cleaned
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(cleaned)
# print(vectorizer.get_feature_names())
print(vectorizer.get_stop_words())
print(X.shape)
```
