```python
# Utils
nlp = spacy.load("en_core_web_sm")

def keep(token):
    if token.is_stop: return False
    if not token.is_alpha: return False
    return True
    

def normalize(item):
    doc = nlp(item)
    print(item)
    return " ".join([str(token) for token in doc if keep(token)])
```

```python
MIN_EDIT_DISTANCE_THRESHOLD = 0.8 # 80% of characters should match


edit_distance_dynamic = dict()

def edit_distance(s, t):
    if s == "": return len(t)
    if t == "": return len(s)
    rs = s[:-1]
    rt = t[:-1]
    cost = s[-1] != t[-1]
    
    return min([
        edit_distance(rs, t) + 1, 
        edit_distance(s, rt) + 1, 
        edit_distance(rs, rt) + cost
    ])

def get_dynamic_dist(s, t):
    key = tuple(sorted([s, t]))
    stored = edit_distance_dynamic.get(key, None)
    
    if not stored:
        edit_distance_dynamic[key] = edit_distance(s, t)
        
    return edit_distance_dynamic[key]
    

def is_match(t1, t2):
    max_changes = round((1 - MIN_EDIT_DISTANCE_THRESHOLD) * len(t2))
    return get_dynamic_dist(t1, t2) <= max_changes
        
        

def find_closest(food_items, inpt):
    inpt_tokens = normalize(inpt).split()
    
    
    
    match_counts = list()
    
    for item in food_items:
        item_tokens = item.lower().split()
        match = 0
        
        for i1, t1 in enumerate(inpt_tokens):
            index = next((i2 for i2, t2 in enumerate(item_tokens) if is_match(t1, t2)), None)
            
            if index != None:
                item_tokens.pop(index)
                match += 1
                
        match_counts.append((match, -len(item), item))
        
    return max(match_counts)[2]
```

```python
import time
start = time.time()
closest = find_closest(food_items, "smith apple")
end = time.time()
print(closest)
print(end - start)
```
